```
from google.colab.patches import cv2_imshow #Colab'de görüntünün okunmasını sağlayan modül
import numpy as np #Matematiksel olarak dizide çalışmayı sağlayan numpy kütüphanesi
import cv2 #Görüntü üzerinde işlemler yapmayı sağlayan OpenCV kütüphanesi
import glob #Yol isminin alınmasını sağlayan kütüphane
#Görüntü klasörünün yerini belirtir.
path = "/content/drive/My Drive/Colab Notebooks/Dataset/Button/Rastgele/*.*"
for file in glob.glob(path):
  #print(file)
  image= cv2.imread(file,0) #Grayscale
  #image= cv2.imread(file) #RGB görüntü
  cv2_imshow(image)
  #print(image)
  #Görüntü İşleme Aşamaları
  # 1- RGB,HSV ve LAB kanallarının kontrol edilmesi
  #image= cv2.cvtColor(image, cv2.COLOR_BGR2LAB )
  #print('LAB Values:',image)
  #print(LABimage)
  #Grayimage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  gaussianTH = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,25,15)
  meanTH = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
           cv2.THRESH_BINARY,25,15)
  print('Gaussian Threshold:')
  cv2_imshow(gaussianTH)
  print('Mean Threshold:')
  cv2_imshow(meanTH)

  #Canny Edge
  canny=cv2.Canny(image,100,200)
  cv2_imshow(canny)
```
